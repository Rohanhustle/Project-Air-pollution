```python
import matplotlib.pyplot as plt
from skimage.io import imread_collection
import pandas as pd
import numpy as np
from sklearn.externals import joblib
```

```python
mfinal=pd.read_csv('datafinal.csv')
mfinal=mfinal.drop(['Unnamed: 0'],axis=1)
mfinal
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

xp=mfinal[['RED','GREEN','BLUE','CONTRAST','ENTROPY']]
yp=mfinal['PM']

x_train,x_test,y_train,y_test=train_test_split(xp,yp,test_size=0.6,random_state=10)

reg=LinearRegression()

reg.fit(x_train,y_train)

reg.predict(x_test)
reg.score(x_test,y_test)
```

```python
gr=pd.read_csv('graphtesting.csv')

gr
```

```python
%matplotlib inline
plt.figure(figsize=(10,5))
plt.xlabel('RED')
plt.ylabel('PM')
plt.scatter(mfinal['RED'],mfinal['PM'],color='red',marker='+')
plt.plot(mfinal['RED'],reg.predict(gr),color='blue')
```
